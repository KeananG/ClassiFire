```python
# take files and seperates .json and .csv files
def import_list(files, path=''):
    '''
    Function takes a list of strings, seperates .json and .csv, option to add a path name to all files
    files : list of files .csv or .json
    path : input path name if file is not in the same directory
    '''
    file_list = files
    json_files = []
    csv_files = []
    for file_name in file_list:
        if file_name.endswith(".json"):
            json_files.append(path + file_name)

        elif file_name.endswith(".csv"):
            csv_files.append(path + file_name)
            
    return json_files, csv_files

json_files, csv_files = import_list(files)
# Add path name
```

# 2. Create a pandas multilevel index DataFrame

```python
def Panda(data,date=None):
    '''
    Function: load in multiple json files, clean and create multilevel index dataframe, run with parallel process
                output is a list of multilevel index dataframes 
                
    data : input list of json file paths
    
    date : used for creating a single level of (will drop and run in new function)
    
    
    Summary:
    
    threshold = file of each stations latest record, when pulling data, post request form pulled from 2023-2014
                so if latest record as 2019, then all rows after are -9999, except for date, year, and day_of_year
    
    key_name = list of station code names
    values = station data
    
    create rows, header, dataset: 
            removes the header and pulls the actual data, 
            input headers are incorrect so header is created with list below
            
    creating dataframes:
            variable name for each df is the station code name, threshold is used to drop all rows that are beyond
            collection date, '-9999' is changed to nulls, and any rows missing all weather metrics have their 
            date set to null, this ensures that day will not be used.
            
    final step: concats all dataframes together with multilevel indexing, each station at index 0, data at index 1
    '''
    # Station latest recorded year, stored as dict
    with open('Data/threshold_year.pickle', 'rb') as file:
        threshold = pickle.load(file)
    
    conl = []
    # Reading json files, each contains x number of stations
    with open(data) as file:   
        data = json.load(file)
    for station in data:
        # pull station code name
        key_name = list(station.keys())[0]
        # Get the value of the dictionary
        values = list(station.values())[0]

        # create rows
        rows = [line.strip().split() for line in (list(station.values())[0]).strip().split('\n')]

        header = ['date', 'year', 'day_of_year', 'day_of_run', 'total_solar_radiation_ly',
                  'ave_mean_wind_speed_mph', 'ave_mean_wind_direction_deg', 'max_maximum_wind_gust_mph',
                  'ave_average_air_temperature_deg_f', 'max_average_air_temperature_deg_f',
                  'min_average_air_temperature_deg_f', 'ave_average_relative_humidity',
                  'max_average_relative_humidity', 'min_average_relative_humidity', 'total_precipitation_in']


        # remove header and ending rows
        dataset = rows[7:-6]

        # Creating individual dataframes
        globals()[key_name] = pd.DataFrame(dataset, columns=header)

        # Change year to int so it can be dropped
        globals()[key_name]['year'] = globals()[key_name]['year'].astype(int)

        # Dropping the extra rows
        globals()[key_name] = globals()[key_name].drop(globals()[key_name][globals()[key_name]['year'] > threshold[key_name]].index)

        # Changing all -9999 to nulls
        globals()[key_name]= globals()[key_name].replace('-9999', np.nan)

        # Removing all rows that have all station metrics as nulls
        for index, row in globals()[key_name].iterrows():
            if row.iloc[-11:].isna().all():
                # use .at, might be faster
                globals()[key_name].at[index, 'date'] = np.nan


        # Creating multi level dataframe
        globals()[key_name].columns = pd.MultiIndex.from_product([[key_name], globals()[key_name].columns])

        # create dataframe of only the dates
        if date:
            globals()[key_name] = globals()[key_name].xs('date', axis = 1, level = 1)
            
        # Creating list of dataframes to concat
        conl.append(globals()[key_name])

    final_dfj = pd.concat(conl, axis = 1)

    return final_dfj

```

```python
# Run panda() function in parallel
results = Parallel(n_jobs=-1, verbose=1)(
    delayed(Panda)(files) for files in json_files
)
```

```python
json_df = pd.concat(results, axis=1)
```

```python
def null_checker(data):
    '''
    Checks if -9999 value is in daataframe, converts columns to numeric to ensure higher accuracy 
    Inputs:
    data : DataFrame to check, single level index
    '''
    # data copy and convert columns to float
    js = data.copy()
    js.drop(columns='date', inplace=True)
    js =js[js.select_dtypes(include='object').columns].astype(float)
    # null value to find
    value = -9999
    is_present = (js[-10:].values == value).any()
    count = (js[-10:].values == value).sum()

    if is_present:
        print(f"The value is present in the DataFrame.\n{count}")
    else:
        print("The value is not present in the DataFrame.")
        
null_checker(json_df['WGOL'])
```

```python
# Number of stations
def number_of_stations(df, Return=True):
    '''
    Returns the number of columns in a multilevel index, also returns a list of the columns
    df : multilevel index dataframe
    Return : True  - set to False to only return nothing and only print
    '''
    col = df.columns.get_level_values(0).unique().tolist()
    print(f'Number of Stations: {len(col)}')
    if Return:
        return col

station_names = number_of_stations(json_df)
```

```python
# station index locator
def station_name_locator(station):
    '''
    Function locates the index of a station in the list returned by number_of_stations
    Used to find index location of certain stations in stations_names list
    '''
    counter = -1
    for s in station_names:
        counter +=1
        if s == station:
            print(counter)
            
station_name_locator('WGOL')
```
